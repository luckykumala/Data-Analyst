```python
Q1TC=df_dummy['TotalCharges'].quantile(0.25)
Q3TC=df_dummy['TotalCharges'].quantile(0.75)
IQRTC=Q3TC-Q1TC
lower_boundTC=Q1TC-1.5*IQRTC
upper_boundTC=Q3TC+1.5*IQRTC
```

```python
Q1MC=df_dummy['MonthlyCharges'].quantile(0.25)
Q3MC=df_dummy['MonthlyCharges'].quantile(0.75)
IQRMC=Q3MC-Q1MC
lower_boundMC=Q1MC-1.5*IQRMC
upper_boundMC=Q3MC+1.5*IQRMC
```

```python
Q1t=df_dummy['tenure'].quantile(0.25)
Q3t=df_dummy['tenure'].quantile(0.75)
IQRt=Q3t-Q1t
lower_boundt=Q1t-1.5*IQRt
upper_boundt=Q3t+1.5*IQRt
```

```python
outliersTC=df_dummy[df_dummy['TotalCharges']>upper_boundTC]
outliersMC=df_dummy[df_dummy['MonthlyCharges']>upper_boundMC]
outlierst=df_dummy[df_dummy['tenure']>upper_boundt]
```
